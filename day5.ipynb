```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x=np.array([2,4,6,8])
y=np.array([81,93,91,97])
```

```python
#데이터 분포를 그래프로 나타냄
plt.scatter(x,y)
plt.show()
```

```python
#기울기 a의 값과 절편 b의 값을 초기화
a=0
b=0
```

```python
#학습률 결정
lr = 0.03
```

```python
#몇번 반복될지 설정 
epochs = 2001
```

```python
# x값이 총 몇개인지 센다
n=len(x)
```

```python
# 경사 하강법을 시작
for i in range(epochs):
    y_pred = a*x + b
    error = y - y_pred

    a_diff = (2/n)*(sum(-x*error))
    b_diff = (2/n)*(sum(-error))

    a = a - lr*a_diff
    b = b - lr*b_diff

    if i % 100==0:
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i,a,b))
```

```python
#앞에서 구한 a,b로 그래프 그리기
y_pred = a*x + b
plt.scatter(x,y)
plt.plot(x, y_pred,'r')
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x=np.array([[2,0],[4,4],[6,2],[8,3]])
y=np.array([81,93,91,97])

model = Sequential()

#입력 변수가 2개 이므로 input_dim에 2를 입력합니다.
model.add(Dense(1,input_dim=2,activation='linear'))
model.compile(loss='mse',optimizer='sgd')

model.fit(x,y,epochs=2000)
```

```python
#임의의 학습 시간과 과외 시간을 집어넣어 점수를 예측하는 모델을 테스트해 보겠습니다. 
hour = 7
private_class = 4
prediction = model.predict([[hour,private_class]])

print("%.f시간을 공부하고 %.f 시간의 과외를 받을 경우,예상점수는 %.02f 점입니다"
%(hour,private_class,prediction))
```
